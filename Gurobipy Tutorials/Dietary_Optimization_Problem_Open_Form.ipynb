```python
#Importing necessary libraries and initializing the model 
from gurobipy import *
import numpy as np
import gurobipy as gp
from gurobipy import GRB
from gurobipy import quicksum

mdl = gp.Model("Diyet Problemi")
```

```python
#Defining variables
x1 = mdl.addVar(vtype= GRB.INTEGER, name='kek')
x2 = mdl.addVar(vtype= GRB.INTEGER, name='dondurma')
x3 = mdl.addVar(vtype= GRB.INTEGER, name='kola')
x4 = mdl.addVar(vtype= GRB.INTEGER, name='pasta')
```

```python
#Constraints
contraint1 = mdl.addConstr(400*x1 + 200*x2 + 150*x3 + 500*x4 >= 500, name='Kalori')
constraint2 = mdl.addConstr(3*x1 + 2*x2 + 0*x3 + 0*x4 >= 6, name='ikolata')
constraint3 = mdl.addConstr(2*x1 + 2*x2 + 4*x3 + 4*x4 >= 10, name='eker')
constraint4 = mdl.addConstr(2*x1 + 4*x2 + 1*x3 + 5*x4 >= 8, name='Ya')
```

```python
#Setting Objective Function
mdl.setObjective(50*x1 + 20*x2 + 30*x3 + 80*x4, sense= GRB.MINIMIZE)
```

```python
#Optimize it!
mdl.optimize()
```
